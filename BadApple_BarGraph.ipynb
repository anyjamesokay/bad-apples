```python
import numpy as np
import matplotlib
matplotlib.use('agg')  #Most important line of the entire code. Makes sure matplotlib doesn't lag out. Credits to DanielGoldFarb on Github
import matplotlib.pyplot as plt
import cv2 as cv
import os
from PIL import Image
main_dir = r"C:\Users\ASPIRE\Desktop\BadApple\frames"
sample = r"C:\Users\ASPIRE\Desktop\BadApple\frames\output_0250.jpg"
sample1 = r"C:\Users\ASPIRE\Desktop\BadApple\frames\output_3000.jpg"
target = r"C:\Users\ASPIRE\Desktop\BadApple\frames_output"
rng = np.random.Generator(np.random.PCG64())
```

```python
def convert_2d(image): #turns RGB frame into a 2d numpy array of 1s and 0s, without OpenCV...
    twod = np.zeros((image.shape[0], image.shape[1]))
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            if np.average(image[i][j]) >= 128:
                twod[i][j] = 1
    return twod
#Generate the bar now. 2 datasets per row, for 36 rows. A crude way to do it, but it worked
def generateBar(image, path_destination, colorlist, k):
    l = list("aAbBcCdDeEfFgGhHiIjJkKlLmMnNoOpPqQrRsStTuUvVwWyY") 
    x, y = np.shape(image)[1], np.shape(image)[0]
    global_p = np.zeros(x)
    p1, p2 = None, None
    for i in range(1, y):
        p1 = np.array(image[y-i])
        p2 = np.array([abs(q-1) for q in p1])
        plt.bar(l, p1, bottom = global_p, color=colorlist[0])
        plt.bar(l, p2, bottom = global_p, color=colorlist[1])
        global_p = global_p + p1 + p2

    plt.title("Bad Apple!! but it's a bar graph")
    plt.savefig(fr'{path_destination}\image{k:04}.png')
    plt.close('all')
```

```python
#For testing how specific frames would look like. Choose frames with lots of gray or lots of chaotic shapes.
a = np.asarray(Image.open(sample))
b = np.asarray(Image.open(sample1))
k = convert_2d(a)
l = convert_2d(b)
#This might not be the best interpolation choice, definitely experiment with others.
lres = cv.resize(k, None, fx=0.1, fy=0.1, interpolation=cv.INTER_NEAREST) 
mres = cv.resize(l, None, fx=0.1, fy=0.1, interpolation=cv.INTER_NEAREST)
fig, ax = plt.subplots(1, 2)
ax[0].imshow(k)
ax[1].imshow(lres)
x, y = np.shape(lres)[1], np.shape(lres)[0]
print(x, y)
print(lres[y-1])
```
