```python
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image
directory = r"C:\Users\ASPIRE\Desktop\BadApple\frames\output_1700.jpg"
rng = np.random.Generator(np.random.PCG64())
```

```python
img = Image.open(directory)
a = np.asarray(img)
def convert_2d(image):
    twod = np.zeros((image.shape[0], image.shape[1]))
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            if np.sum(image[i][j]) >= 128*3:
                twod[i][j] = 1
            else:
                twod[i][j] = 0
    return twod

def relative_density(image):
    dens = np.count_nonzero(image)
    return -0.45*round(dens/np.size(image), 6)+0.6

def hoshen_kopelman(frame):
  frame = convert_2d(frame)
  prob = relative_density(frame)
  size1 = frame.shape[0]
  size2 = frame.shape[1]
  lattice = rng.random((size1, size2))
  lattice = (lattice < prob)*frame 
  label = 1
  labelsheet = np.zeros((size1, size2))

  for i in range(size1):
    for j in range(size2):
      if lattice[i][j] == 1:
        if labelsheet[i][j-1] == 0 and labelsheet[i-1][j] == 0: #Rule 1
          labelsheet[i][j] = label
          label += 1

        if labelsheet[i][j-1] != 0 and labelsheet[i-1][j] == 0: #Rule 2
          labelsheet[i][j] = labelsheet[i][j-1]

        if labelsheet[i][j-1] == 0 and labelsheet[i-1][j] != 0:
          labelsheet[i][j] = labelsheet[i-1][j]

        if labelsheet[i][j-1] != 0 and labelsheet[i-1][j] != 0: #Rule 3
          labelsheet[i][j] = labelsheet[i-1][j]
          labelsheet[labelsheet == labelsheet[i][j-1]] = labelsheet[i-1][j]

  return labelsheet
```

```python
labelsheet = hoshen_kopelman(a)
if np.sum(labelsheet) == 0:
    plt.imshow(labelsheet, cmap="plasma", vmin = 0, vmax = 1)
    plt.colorbar()
    plt.title("Percolation of Frame")
else:
    plt.imshow(labelsheet, cmap="plasma")
    plt.colorbar()
    plt.title("Percolation of Frame")
```
